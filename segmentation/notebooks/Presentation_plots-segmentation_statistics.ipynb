```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
out_fname = 'segmentation_party_statistics.pickle'
df = pd.read_pickle(out_fname)
```

```python
neuron_names = df.index
for i, name in enumerate(neuron_names):
    row = df.loc[name,:]#.to_numpy()
    
#     plt.violinplot(row, positions=i*np.ones_like(row))
    sns.violinplot(row)

    x = i*np.ones_like(row)
    sns.swarmplot(x=x, data=row)
    
plt.xticks(np.arange(len(df.index))+1)
plt.xlabel(list(df.index))
```

```python

sns.swarmplot(data=df.T, color='white')
sns.violinplot(data=df.T)
plt.xticks(rotation=45)
plt.ylim([0,40])
plt.title("Statistics from manual annotation on ")
```
