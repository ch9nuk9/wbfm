```python
out_fname = 'segmentation_party_statistics.pickle'
df = pd.read_pickle(out_fname)
```

```python
plt.figure(figsize=(15,5))
plt.plot(df.loc['Total'], 'o')
plt.ylabel('Number of neurons')
plt.xticks(ticks=range(len(df.columns)), labels=list(map(lambda x: x.split('-')[1], list(df.columns))))
plt.xticks(rotation=45);
plt.xlabel("Dataset name (same name means different frames)")
plt.title("Number of neurons with different datasets");
```

```python
df_normalized = (df.T)[['False positives', 'False negatives', 'Undersegmentation', 'Oversegmentation']].div(df.T['Total'], axis=0)

plt.figure(figsize=(15,5))
sns.swarmplot(data=df_normalized, color='white')
sns.violinplot(data=df_normalized)
# plt.xticks(rotation=45)
plt.ylim([0,0.25])
plt.ylabel("Fraction of total")
plt.title("Statistics from manual annotation (15 volumes)");
```
