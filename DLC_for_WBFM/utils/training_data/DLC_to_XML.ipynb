```python
from lxml import etree as ET
import pandas as pd
from xml_utils import xml_pretty_print
```

# Importing DLC

```python
dlc_fname = 'CollectedData_Charlie.h5'

df = pd.read_hdf(dlc_fname)
df = df['Charlie']
```

```python
df
```

```python
df = df.sort_index() # Filenames may be a different order

all_neurons = df.columns.levels[0]
print(all_neurons)
all_files = df.index
print(all_files)
```

```python
df
```

# Making XML

```python
# Initialize
xm = ET.Element("root")
xm1 = ET.SubElement(xm, "trackfile", version="1")
```

```python
# Add a neuron

for i, this_neuron in enumerate(all_neurons):
    trackgroup = ET.SubElement(xm1, "trackgroup", description="Neuron {} from DLC".format(i))
    t = ET.SubElement(trackgroup, "track", id=str(i))
    for i_t, this_file in enumerate(all_files):
        x = str(df[this_neuron]['x'][this_file])
        y = str(df[this_neuron]['y'][this_file])
        z = str(df[this_neuron]['z'][this_file])
        ET.SubElement(t, "detection",
                      classname="plugins.nchenouard.spot.Detection",
                      color=str(i),
                      t=str(i_t),
                      type="1",
                      x=x, y=y, z=z)
```

```python
xml_pretty_print(xm, max_depth=3)
```

```python
# SAVE
tree = ET.ElementTree(xm)

tree.write('test_XML_from_DLC.xml', pretty_print=True)#, xml_declaration=True, encoding="utf-8")
```
